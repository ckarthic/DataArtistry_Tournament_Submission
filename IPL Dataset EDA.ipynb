```python
df = pd.read_json('https://gramener.com/datasets/ipl-may-2016/matches?_limit=600&_format=json&_meta=y')
#df.head()
```

```python
print(df.head())
print(df.shape)
```

```python
df.loc[df.season == 2015,:].groupby('result').sum()
```

```python
for (method, group) in df.groupby('season'):
    print("{0} shape={1}".format(method, group.shape))
print(df.loc[df.win_by_wickets == 0,:].groupby('season')['win_by_runs'].describe())
print(df.loc[df.win_by_runs == 0,:].groupby('season')['win_by_wickets'].describe())
```

```python
for (method, group) in df.groupby('city'):
    print("{0} shape={1}".format(method, group.shape))
print(df.loc[df.win_by_wickets == 0,:].groupby('city')['win_by_runs'].describe())
print(df.loc[df.win_by_runs == 0,:].groupby('city')['win_by_wickets'].describe())
```
